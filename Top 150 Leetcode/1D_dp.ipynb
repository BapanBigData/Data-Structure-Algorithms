```python
# 70. Climbing Stairs
# Easy

def climb_stairs(n: int) -> int:
    
    # top-down approach
    def solve_memo(memo, n):
        # base cases
        if (n == 1) or (n == 0):
            return 1
        
        # check if the state is already computed
        if memo[n] is not None:
            return memo[n]
        
        memo[n] = solve_memo(memo, n-1) + solve_memo(memo, n-2)
        
        return memo[n]
    
    # bottom-up approach
    def solve_tabulation(n):
        # initializing a dp array
        dp = [None] * (n+1)
        
        # base cases
        dp[0] = dp[1] = 1
        
        for i in range(2, n+1):
            dp[i] = dp[i-1] + dp[i-2]
        
        return dp[n]
    
    # space optimized bottom-up soln
    def space_opt(n):
        a = 1
        b = 1
        
        for c in range(2, n+1):
            c = a + b
            a = b
            b = c
            
        return b
    
    # initialize the memo array
    memo = [None] * (n+1)
    
    return solve_memo(memo, n), solve_tabulation(n), space_opt(n)

climb_stairs(21)
```

```python
# 198. House Robber
# Medium

def rob(nums: list[int]) -> int:
    n = len(nums)
    
    # top-down approach
    def solve_memo(memo, i):
        # base cases
        if (i >= n):
            return 0
        
        if (i == (n-1)):
            return nums[i]
        
        # check if the state is already computed
        if memo[i] is not None:
            return memo[i]
        
        # incle case
        incl = nums[i] + solve_memo(memo, i+2)
        
        # excl case
        excl = 0 + solve_memo(memo, i+1)
        
        memo[i] = max(incl, excl)
        
        return memo[i]
    
    # bottom-up approach
    def solve_tabulation(n):
        # let's initialize the dp array
        dp = [0] * (n+1)
        
        # base case
        dp[n-1] = nums[n-1]
        
        for i in range(n-2, -1, -1):
            incl = nums[i] + dp[i+2]
            excl = 0 + dp[i+1]
            
            dp[i] = max(incl, excl)
            
        return dp[0]
    
    # space optimized bottom-up soln
    def space_opt(n):
        a = 0
        b = nums[n-1]
        
        for i in range(n-2, -1, -1):
            incl = nums[i] + a
            excl = 0 + b
            
            c = max(incl, excl)
            a = b
            b = c
        
        return b
    
    # initialize the memo array
    memo = [None] * n
    
    return solve_memo(memo, 0), solve_tabulation(n), space_opt(n)

nums = [2,7,9,3,1]
rob(nums)
```

```python
# 139. Word Break
# Medium

def word_break(s: str, wordDict: list[str]) -> bool:
    n = len(s)
    
    def solve_memo(memo, i):
        # base case
        if (i >= n):
            return True
        
        # check if the state is already computed
        if (memo[i] != -1):
            return memo[i]
        
        flag = False
        for word in wordDict:
            l = len(word)
            
            if (i + l) > n:
                continue
            
            if (s[i:(i+l)] == word) and solve_memo(memo, i+l):
                flag = True     
                break   # break, since we found a valid split
        
        memo[i] = flag
        
        return memo[i]
    
    def solve_tabulation(n):
        # initializing the dp array
        dp = [-1] * (n+1)
        
        dp[n] = True
        
        for i in range(n-1, -1, -1):
            
            flag = False
            for word in wordDict:
                l = len(word)
                
                if (i + l) > n:
                    continue
                
                if (s[i:(i+l)] == word) and dp[i+l]:
                    flag = True
                    break   # break, since we found a valid split
                
            dp[i] = flag
        
        return dp[0]
    
    # initializing the memo array
    memo = [-1] * n
    
    return solve_memo(memo, 0), solve_tabulation(n)

s = "applepenapple"
wordDict = ["apple","pen"]

word_break(s, wordDict)
```

```python
# 322. Coin Change
# Medium
import math

def coin_change(coins: list[int], amount: int) -> int:
    
    def solve_memo(memo, amount):
        # base case
        if (amount == 0):
            return 0
        
        # check if the state is already computed
        if memo[amount] is not None:
            return memo[amount]
        
        mn = math.inf
        for coin in coins:
            if (amount - coin) >= 0:
                mn = min(mn, 1 + solve_memo(memo, amount-coin))
        
        memo[amount] = mn
        
        return memo[amount]
    
    # bottom-up soln
    def solve_tabulation(amount):
        # initialize the dp array
        dp = [0] * (amount+1)
        
        for i in range(1, amount+1):
            
            mn = math.inf
            for coin in coins:
                if (i-coin) >= 0:
                    mn = min(mn, 1 + dp[i-coin])
            
            dp[i] = mn 
        
        return dp[amount]
    
    # initializing the memo array
    #memo = [None] * (amount+1)
    # res = solve_memo(memo, amount)
    
    res = solve_tabulation(amount)
    
    return -1 if res == math.inf else res 


coins = [1,2,5]
amount = 11

coin_change(coins, amount)
```

```python
# 300. Longest Increasing Subsequence
# Medium

def lengthOfLIS(nums: list[int]) -> int:
    n = len(nums)
    
    def solve_memo(memo, curr, prev):
        # base case
        if (curr == n):
            return 0
        
        if memo[curr][prev] is not None:
            return memo[curr][prev]
        
        incl, excl = 0, 0
        
        # incl case
        if (prev == -1) or (nums[curr] > nums[prev]):
            incl = 1 + solve_memo(memo, curr+1, curr)
        
        excl = 0 + solve_memo(memo, curr+1, prev)
        
        memo[curr][prev] = max(incl, excl)
        
        return memo[curr][prev]
    
    # bottom-up soln
    def solve_tabulation(n):
        # initialize the dp array
        dp = [[0]*(n+1) for _ in range(n+1)]
        
        for curr in range(n-1, -1, -1):
            for prev in range(curr-1, -2, -1):
                
                incl, excl = 0, 0
                
                if (prev == -1) or (nums[curr] > nums[prev]):
                    incl = 1 + dp[curr+1][curr]
                
                excl = 0 + dp[curr+1][prev]
                
                dp[curr][prev] = max(incl, excl)
        
        return dp[0][-1]
    
    # bottom-up space optimized soln
    def space_opt(n):
        #let's initialize two array
        row_0 = [0] * (n+1)
        row_1 = [0] * (n+1)
        
        for curr in range(n-1, -1, -1):
            for prev in range(curr-1, -2, -1):
                
                incl, excl = 0, 0
                
                if (prev == -1) or (nums[curr] > nums[prev]):
                    incl = 1 + row_0[curr]
                
                excl = 0 + row_0[prev]
                
                row_1[prev] = max(incl, excl)
            
            row_0 = row_1[:]
        
        return row_0[-1]
    
    # initialize the memo array
    memo = [[None] * n for _ in range(n)]
    
    return solve_memo(memo, 0, -1), solve_tabulation(n), space_opt(n)

nums = [10,9,2,5,3,7,101,18]
lengthOfLIS(nums)
```
