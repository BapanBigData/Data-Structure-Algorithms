```python
import math
# 45. Jump Game II

# You are given a 0-indexed array of integers nums of length n. You are initially positioned at nums[0].
# Each element nums[i] represents the maximum length of a forward jump from index i. In other words, if you are at nums[i], 
# you can jump to any nums[i + j] where:
# 0 <= j <= nums[i] and
# i + j < n
# Return the minimum number of jumps to reach nums[n - 1]. 
# The test cases are generated such that you can reach nums[n - 1].

def minJump(nums: list):
    
    def top_down_solver(memo, i, n):
        if i >= n:
            return math.inf
        
        ## base case
        if i == n-1:
            memo[i] = 0
            return memo[i]
        
        ## check if the state is already computed
        if memo[i] is not None:
            return memo[i]
        
        ans = math.inf
        for jump in range(1, nums[i]+1):
            f = 1 + top_down_solver(memo, i+jump, n)
            ans = min(ans, f)

        ## store the computation
        memo[i] = ans

        return memo[i]
    
    def bottom_up_solver(n):
        dp = [None]*n

        ##  base case
        dp[n-1] = 0

        for i in range(n-2, -1, -1):
            ans = math.inf
            for jump in range(1, nums[i]+1):
                if (i + jump) <= (n-1):
                    f = 1 + dp[i+jump]
                    ans = min(ans, f)
            
            dp[i] = ans
        
        return dp[0]

    
    n = len(nums)
    # memo = [None] * n

    # return top_down_solver(memo, 0, n)

    return bottom_up_solver(n)


nums = [2, 3, 1, 1, 4]
res = minJump(nums)
print(res)
```

```python
## 46. Permutations
## Given an array nums of distinct integers, return all the possible permutations. You can return the answer in any order.

def permutations(nums: list):

    def solver(start, lst, n, res):
        if start == n:
            res.append(lst[:])
            return
        
        for i in range(start, n):
            lst[i], lst[start] = lst[start], lst[i]
            solver(start+1, lst, n, res)

            ## backtrack
            lst[i], lst[start] = lst[start], lst[i]
        
        return
    
    n = len(nums)
    res = []

    solver(start=0, lst=nums, n=n, res=res)
    return res

nums = [1, 2, 3, 4]
res = permutations(nums)
print(res)
```
