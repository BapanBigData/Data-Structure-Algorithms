```python
## Binary Search: Given an sorted array and an element x,
## get the index of x in the array if not present reture -1

def binary_search(nums: list[int], x: int):
    n = len(nums)
    s, e = 0, n-1
    
    while (e >= s):
        
        mid = (s + e) // 2
        
        if (nums[mid] == x):
            return mid
        
        if nums[mid] > x:
            e = mid - 1
        else:
            s = mid + 1
            
    return -1

nums = [1, 2, 3, 4, 5, 6]
x = 5
print(binary_search(nums, x))
```

```python
## first occurrence

def fisrt_occurrence(nums: list[int], x: int):
    n = len(nums)
    s, e = 0, n-1
    
    while (e >= s):
        mid = (s + e) // 2
        
        if (nums[mid] == x):
            if ((mid-1) >= 0) and (nums[mid-1] == nums[mid]):
                e = mid - 1
            else:
                return mid
            
        elif (nums[mid] > x):
            e = mid -1
        else:
            s = mid + 1
    
    return -1

nums = [1, 1, 2, 2, 2, 2, 3]
x = 2

fisrt_occurrence(nums, x)
```

```python
## last occurrence

def last_occurrence(nums: list[int], x: int):
    n = len(nums)
    s, e = 0, n-1
    
    while (e >= s):
        mid = (s + e) // 2
        
        if (nums[mid] == x):
            if ((mid+1) < n) and (nums[mid+1] == nums[mid]):
                s = mid + 1
            else:
                return mid
        
        elif (nums[mid] > x):
            e = mid - 1
        else:
            s = mid + 1
    
    return -1

nums = [1, 1, 2, 2, 2, 2, 3]
x = 2

last_occurrence(nums, x)
```

```python
## number of occurrence

def cnt_occurrence(nums: list[int], x: int):
    first_occ = fisrt_occurrence(nums, x)
    last_occ = last_occurrence(nums, x)
    
    return 0 if (first_occ == -1 or last_occ == -1) else last_occ - first_occ + 1

nums = [1, 1, 2, 2, 2, 2, 3]
x = 4

cnt_occurrence(nums, x)
```

```python
## search in rotated sorted array

def search(nums: list[int], target: int):
    n = len(nums)
    s, e = 0, n-1
    
    while (e >= s):
        mid = (s + e) // 2
        
        if (nums[mid] == target):
            return mid
        
        ## check in which line the mid is falling
        ## mid falling on the first increasing line
        if (nums[s] <= nums[mid]):
            if (nums[s] <= target < nums[mid]):
                e = mid - 1
            else:
                s = mid + 1
        
        ## mid falling on the second increasing line
        elif (nums[e] >= nums[mid]):
            if (nums[mid] < target <= nums[e]):
                s = mid + 1
            else:
                e = mid - 1
            
    return -1

nums = [5, 1, 3]
x = 5

search(nums, x) 
```
